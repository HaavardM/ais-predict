```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2
from ais_predict.visualization import plotting
sns.set_context("paper")
sns.set()

latex_width = 369.88583
plt.rcParams["figure.figsize"] = plotting.set_size(latex_width, 0.6)

tex_fonts = {
    # Use LaTeX to write all text
    "text.usetex": True,
    "font.family": "serif",
    # Use 10pt font in plots, to match 10pt font in document
    "axes.labelsize": 10,
    "font.size": 10,
    # Make the legend/label fonts a little smaller
    "legend.fontsize": 8,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8,
    "text.latex.preamble": r'\usepackage{amsfonts}'
}

plt.rcParams.update(tex_fonts)

name = "straight_line"
default_limits = (-200, 500)

df = pd.read_pickle(f"{name}.pkl")
```

```python
len(df.test_id.unique())
```

```python
import seaborn as sns
from scipy.interpolate import interp1d
d = df.copy()
d["traj_err"] = d.trajectory_error.apply(lambda x: np.linalg.norm(x[:2])).copy()
d["path_err"] = d.path_error.apply(lambda x: np.linalg.norm(x[:2])).copy()
d = d[["test_id", "timestep", "method", "y_source", "traj_err", "path_err", "nees", "train_samples"]]
d = d.loc[d.timestep > 0]
result_df = []
pred_t = np.arange(5, 30, step=5) * 60
pred_traj_err = np.ones_like(pred_t) * np.nan
pred_path_err = np.ones_like(pred_t) * np.nan
for (_, g) in d.groupby(["method", "y_source", "test_id"]):
    t = g.timestep.to_numpy()
    t_msk = (pred_t < t[-1]) & (pred_t > t[0])
    traj_err = g.traj_err.to_numpy()
    path_err = g.path_err.to_numpy()
    nees = g.nees.to_numpy()
    traj_f = interp1d(t, traj_err)
    path_f = interp1d(t, path_err)
    nees_f = interp1d(t, nees)
    temp = pd.DataFrame({
        "time": pred_t[t_msk],
        "traj_err": traj_f(pred_t[t_msk]),
        "path_err": path_f(pred_t[t_msk]),
        "nees": nees_f(pred_t[t_msk])
    })
    temp["method"] = g.method.iloc[0]
    temp["y_source"] = g.y_source.iloc[0]
    temp["test_id"] = g.test_id.iloc[0]
    temp["train_samples"] = g.train_samples.iloc[0]
    result_df.append(temp)
result_df = pd.concat(result_df)
    

```

```python
r = result_df.copy()
r.time = r.time.divide(60).astype(np.int)
r.method = r.method.map({
    "dyngp_pdaf": "GP-EKF w/ PDAF",
    #"dyngp_pdaf_no_P": "GP-EKF w/ MOD-PDAF",
    "dyngp_syn": "GP-EKF w/ SL",
    #"dyngp_syn_no_P": "GP-EKF w/ MOD-SL", 
    "dyngp_no_pdaf": "GP-EKF",
    "cvm": "CVM",
    "posgp": "Direct GP"
})
r.y_source = r.y_source.map({
    "cog": "COG/SOG from AIS",
    "finite_difference": "Finite Difference",
    "position": "Position"
})
r = r.groupby(["time", "method", "y_source"]).path_err.agg([np.mean, np.median]).round().astype(np.int)
r.columns = ["Mean", "Median"]
r = r.unstack(0)
r.columns.names = ["Summary Type", "Time [Minutes]"]
r.columns.name = "Time [Minutes]"
r.index.names = ["Method", "Training Source"]
r = r.stack(0)
r = r.swaplevel(0).swaplevel(1, 2).sort_index()
r.index.names = ["Summary", "Method", "Training Source"]
r
print(r.to_latex())
```

```python
r = result_df.copy()
r.time /= 60
r.time = r.time.astype(np.int)
r.time = r.time.astype("category")
r.method = r.method.map({
    "posgp": "Direct GP",
    "dyngp_no_pdaf": "GP-EKF",
    "cvm": "CVM",
}).astype("category")
width, heigth = plotting.set_size(latex_width)
fig, ax = plt.subplots(1, 3, sharey=True, sharex=True, figsize=(width, 2*heigth / 3))
t = sns.boxplot(y="traj_err", x="time", data=r.loc[(r.method=="GP-EKF") & (r.y_source == "finite_difference")],ax=ax[0], showfliers=False, palette="Blues")
t.set(ylabel="Trajectory Error [m]", xlabel="Time [Minutes]", title=f"GP-EKF")
t = sns.boxplot(y="traj_err", x="time", data=r.loc[r.method=="Direct GP"],ax=ax[1], showfliers=False, palette="Blues")
t.set(ylabel="", xlabel="Time [Minutes]", title="Direct GP")
t = sns.boxplot(y="traj_err", x="time", data=r.loc[r.method=="CVM"], ax=ax[2], showfliers=False, palette="Blues")
t.set(ylabel="", xlabel="Time [Minutes]", title="CVM")
plt.savefig(f"{name}/stats/both_vs_cvm.pdf", format="pdf", bbox_inches='tight')
    #plt.title((g.method.iloc[0] + "_" + g.y_source.iloc[0]).replace("_", "\_"))
```
